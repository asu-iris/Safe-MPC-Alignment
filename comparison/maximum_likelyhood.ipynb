```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
threshold_arr =np.arange(-0.1,-1.0,-0.1)
error_thresh = np.load('../Data/compare/comp_2_err_thresh.npy')
error_thresh_cut = np.load('../Data/compare/comp_2_err_thresh_cut.npy')
std_thresh = np.load('../Data/compare/comp_2_std_thresh.npy')
std_thresh_cut = np.load('../Data/compare/comp_2_std_thresh_cut.npy')
```

```python
plt.figure()
#plt.title("Parameter Error",fontsize=20)
plt.ylabel("Parameter Error",fontsize=25)
plt.xlabel("$\epsilon_g$",fontsize=25)
#plt.ylim(0.0,0.3)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.plot(-threshold_arr,error_thresh,label="Maximum Likelyhood")
plt.errorbar(-threshold_arr, error_thresh, yerr=np.array(std_thresh)/2, capsize=3, fmt="none", ecolor = "black",markersize = 5)
plt.plot(-threshold_arr,error_thresh_cut,label="Ours")
plt.errorbar(-threshold_arr, error_thresh_cut, yerr=np.array(std_thresh_cut)/2, capsize=3, fmt="none", ecolor = "black",markersize = 2)
plt.tight_layout()
plt.grid()
plt.legend(fontsize=20)
plt.savefig('../Data/compare/baseline_2_epsilon.jpg')
plt.show()
```
